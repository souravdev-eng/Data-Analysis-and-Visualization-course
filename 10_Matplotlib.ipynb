## Bar Plots

```python
plants = ["Spinach", "Turnip", "Rhubarb", "Broccoli", "Kale"]
died = [10, 25, 5, 30, 21]
germinated = [74, 88, 56, 69, 59]
```

```python
plt.figure(figsize=(6,4))
plt.bar(plants, germinated, width=0.5)
plt.title("Germination Rate");
```

```python
plt.figure(figsize=(6,4))
plt.bar(plants, died, width=0.5)
plt.title("Germination Rate");
```

```python
plt.figure(figsize=(6,4))
plt.bar(plants, died, color="red")
plt.bar(plants, germinated, width=0.4, color="green", align="edge")
plt.title("Germination Rate");
```

```python
plt.figure(figsize=(6,4))
plt.bar(plants, died, color="red", bottom=germinated, label="Failed Seedlings")
plt.bar(plants, germinated, color="green", label="Germinated Seedlings")
plt.legend();
```

```python
plt.figure(figsize=(6,4))
plt.barh(plants, died);
```

## Histagrams

```python
rand_nums = np.random.randn(100)
plt.figure(figsize=(6,4))
plt.hist(rand_nums*3)
plt.hist(rand_nums*2);
```

## Scatter Plots

```python
height = np.random.randn(50)
f_weights = np.random.randn(50)
m_weights = np.random.randn(50)
```

```python
plt.scatter(f_weights, height, c="blue")
plt.scatter(m_weights, height, c="orange")
```

## Pie Charts

```python
labels = ["Rice", "Roti", "Chicken", "Eggs", "Panir"]
prices = [25.99, 3.24, 9.50, 6.90, 7.42]
```

```python
plt.pie(prices, labels=labels, autopct="%1.1f%%",explode=(0,0,0.1,0,0));
```

## Subplots

```python
plt.figure(figsize=(10,4))
plt.suptitle("Global Fig Title")
plt.subplot(1,3,1)
plt.title("1st")
plt.plot(nums, nums)

plt.subplot(1,3,2)
plt.title("2nd")
plt.plot(nums, nums*nums, c="orange")

plt.subplot(1,3,3)
plt.title("3rd")
plt.plot(nums, nums**3, c="green")
plt.tight_layout(); # For adjust the title
```

```python
plt.figure(figsize=(8,8))
plt.suptitle("Global Fig Title")
plt.subplot(2,2,1)
# Line plot
plt.title("1st")
plt.plot(nums, nums)

# Histagrame
plt.subplot(2,2,2)
plt.title("2nd")
plt.hist(rand_nums*3)
plt.hist(rand_nums*2);

# Scatter Plot
plt.subplot(2,2,3)
plt.title("3rd")
plt.scatter(f_weights, height, c="blue")
plt.scatter(m_weights, height, c="orange")

# Pie Charts
plt.subplot(2,2,4)
plt.title("4th")
plt.pie(prices, labels=labels, autopct="%1.1f%%",explode=(0,0,0.1,0,0));
plt.tight_layout(); # For adjust the title
```
