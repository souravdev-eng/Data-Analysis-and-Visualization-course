### 6. loc

The `loc` method accesses data in a Pandas DataFrame using labels or boolean conditions.

When using loc, you can specify the row labels and column labels you want to select. For example, you could use `df.loc[3, 'col1']` to access the value in the row with label 3 and the column with label "col1". You can also use slices to select multiple rows or columns at once, such as `df.loc[1:5, 'col2':'col4']`.

`loc` can also be used with boolean conditions. For example, you could use `df.loc[df['col1'] > 5, 'col2']` to select the values in the "col2" column where the corresponding value in the "col1" column is greater than 5.

In summary, loc accesses data by using labels to identify rows and columns or boolean conditions to select subsets of rows and columns.

```python
countries.loc["India"]
```

```python
# To display the data into a DataFrame
countries.loc[["India"]]
```

```python
countries.sort_index(inplace=True)
```

```python
countries
```

```python
# Show a slice of array values
countries.loc["Denmark": "India"]
```
