```python
import pandas as pd

houses = pd.read_csv("data/kc_house_data.csv")
titanic = pd.read_csv("data/titanic.csv")
netflix = pd.read_csv("data/netflix_titles.csv", sep="|", index_col=0)
```

## `between()`

```python
houses[houses["bedrooms"].between(5, 7)]
```

## `isin()`

```python
netflix[netflix["country"].isin(["India", "Japan"])].head()
```

## `&` (and oparetor)

```python
women = titanic["sex"] == 'female'
died = titanic["survived"] == 0

titanic[women & died]
```

```python
houses.head()
```

```python
houses[(houses["waterfront"] == 1) & (houses["price"] < 500000)]
```

## `|` (or oparator)

```python
houses[(houses["yr_built"] > 2010) | (houses["yr_built"] <= 2012)]
```

## bitwise nabigation

```python
df = titanic.head()
women = df["sex"] == 'female'
df[~women] # It's turn all data oposite 
```
